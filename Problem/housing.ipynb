```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DataSets/housing.csv')
df.head()
```

```python
# Select only numerical columns for correlation analysis
numeric_df = df.select_dtypes(include=['int64', 'float64'])

print("Numeric Columns:")
print(numeric_df.columns)
```

```python
pearson_corr = numeric_df.corr(method='pearson')
print("Pearson correlation matrix:")
print(pearson_corr)
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(pearson_corr,
            annot=True, # Annotate the heatmap with correlation values
            fmt='.2f', # Format the annotation to 2 decimal places
            cmap='coolwarm', # Use a diverging colormap to show positive and negative correlations. More cmaps: 'viridis', 'plasma', 'inferno', 'magma', 'cividis'
            linewidths=2 # Add lines between cells for better readability
            )
plt.title("Pearson Correlation Heatmap")
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.yticks(rotation=45) # Rotate y-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

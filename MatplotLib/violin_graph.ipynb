```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../DataSets/titanic.csv')
df.head()
```

```python
encoder = LabelEncoder() # Initialize the LabelEncoder
columns_to_encode = ["Embarked", "Sex"]
for column in columns_to_encode:
    df[column] = encoder.fit_transform(df[column])

df_encoded = pd.get_dummies(df) # This will create new columns for each category in the specified columns and fill them with 0s and 1s
print("Encoded DataFrame:")
print(df_encoded.head())
```

```python
sns.violinplot(x="Sex", y="Fare", data=df_encoded)
plt.title("Fare distribution across Gender")
plt.show()
```

```python
sns.violinplot(x="Embarked", y="Age", data=df_encoded)
plt.title("Age distribution across Embarked")
plt.show()
```
